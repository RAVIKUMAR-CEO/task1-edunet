```python
from zipfile import ZipFile
import os

# Step 2: Set correct path
zip_path = '/content/drive/My Drive/Tree_Species_Dataset.zip-20250718T182112Z-1-001.zip'
extract_path = '/content/drive/My Drive/Tree_Species_Dataset'

# Step 3: Check and extract
if not os.path.isfile(zip_path):
    print(f"❌ Error: The path '{zip_path}' does not point to a file.")
else:
    try:
        with ZipFile(zip_path, 'r') as zipObj:
            zipObj.extractall(extract_path)
        print("✅ Extraction completed successfully!")
    except Exception as e:
        print(f"❌ An error occurred during extraction: {e}")
```
